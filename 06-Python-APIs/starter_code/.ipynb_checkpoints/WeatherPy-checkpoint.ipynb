```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

## Perform API Calls

```python
# OpenWeatherMap API Key
api_key = api_keys.api_key

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 

# Build partial query URL
query_url = f"{url}&q="

# set up lists to hold reponse info
lat = []
lng = []
temp = []
hum = []
cloud = []
wind = []
names = []
country = []
date = []
ids = []
x = 1
y = 1


# Loop through the list of cities and perform a request for data on each
print("Beginning Data Retrieval")
print("-------------------------------")

for city in cities:
    # Sleep to help with the call limit
    time.sleep(1)
    
    print("Processing Record " + str(x) + " of Set " + str(y) + "  | " + city)
    print(query_url + city)
    response = requests.get(query_url + city).json()

    try:
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        temp.append(response['main']['temp_max'])
        hum.append(response['main']['humidity'])
        cloud.append(response['clouds']['all'])
        wind.append(response['wind']['speed'])
        date.append(response['dt'])
        country.append(response['sys']['country'])
        ids.append(response['id'])
        names.append(response['name'])
    
    except:
       print("")
    
    if x % 50 == 0: 
        x=1
        y += 1
    else:
        x += 1
        y = y 
        
print("-------------------------------")
print("Data Retrieval Complete")
print("-------------------------------")
        
```

```python
# create a data frame 
weather_dict = {
    "City": names,
    "Cloudiness": cloud,
    "Country": country,
    "Date": date,
    "Humidity": hum,
    "Lat": lat,
    "Lng": lng,
    "Max Temp": temp,
    "Wind Speed": wind
}
weather_data = pd.DataFrame(weather_dict)
weather_data.head()
```

```python
# Export file as a CSV, without the Pandas index
weather_data.to_csv("WorldWeatherData.csv", index=False, header=True)
```

```python
# Latitude vs Temperature Plot
plt.title("City Latitude vs. Max Temperature (01/05/17)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.scatter(weather_data['Lat'], weather_data['Max Temp'] , marker="o", facecolors="blue",\
            edgecolors="black", alpha=0.75)
plt.ylim(0, 120)
plt.xlim(-80, 100)
plt.grid()
plt.savefig("City Latitude vs Max Temperature")
```

```python
# Latitude vs Humidity Plot
plt.title("City Latitude vs. Humidity (01/05/17)")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.scatter(weather_data['Lat'], weather_data['Humidity'] , marker="o", facecolors="blue", edgecolors="black")
plt.ylim(-20, 120)
plt.xlim(-80, 100)
plt.grid()
plt.savefig("City Latitude vs Humidity")
```

```python
# Latitude vs Cloudiness Plot
plt.title("City Latitude vs. Cloudiness (01/05/17)")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
plt.scatter(weather_data['Lat'], weather_data['Cloudiness'] , marker="o", facecolors="blue", edgecolors="black")
plt.ylim(-20, 120)
plt.xlim(-80, 100)
plt.grid()
plt.savefig("City Latitude vs Cloudiness")
```

```python
# Latitude vs Wind Speed Plot
plt.title("City Latitude vs. Wind Speed (01/05/17)")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.scatter(weather_data['Lat'], weather_data['Wind Speed'] , marker="o", facecolors="blue", edgecolors="black")
plt.ylim(-5, 40)
plt.xlim(-80, 100)
plt.grid()
plt.savefig("City Latitude vs Wind Speed")
```

Three observation trends based on the data:

1. Temperature increased as we approach the equator. The highest temperature is near the equator area, then decreases along with the increasing of latitude(both north and south side)

2. Most of the equator area has the humidity around 60%-100%.

3. There is no significant differences of wind speed and cloudiness in cities with different latitude
